# Random Forest


```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import tree


iris=load_iris()
x=iris.data
y=iris.target
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#model train
# n_estimators =number of disiyion tree
# max_samples<n_estimators
RFC=RandomForestClassifier(n_estimators=100,max_depth=4,max_features=3,max_samples=80,criterion="entropy")
RFC.fit(x_train,y_train)
```

```python
selected_tree=RFC.estimators_[0]
tree.plot_tree(selected_tree)
```

```python
selected_tree=RFC.estimators_[5]
tree.plot_tree(selected_tree)
```

```python
y_pred=RFC.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
print(data.isnull().sum)
data.duplicated().sum
```

```python
y=data[["Outcome"]]
x=data.drop("Outcome",axis=1)
print(x)
print(y)
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=10)
print(x_train.shape)
# train the data
RFC=RandomForestClassifier(n_estimators=500,max_depth=4,criterion="entropy")
print(RFC.fit(x_train,y_train))
# predict
y_pred=RFC.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
